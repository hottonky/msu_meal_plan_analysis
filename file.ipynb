```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
swipe_cost = 9.00
meal_plan_cost = 3226
spring_semester_weekdays_2024 = 112
day_of_week = 16
```

1. student meal plan all the way
2. student meal plan + off campus 
    student adds amount spent per week off campus instead of using meal plan (going out with friend)
3. swipes only
4. swipes + off campus
5. off campus spending

```python
# how much would you spend per day outside of the meal plan pn average?
meal_plan_mon = 0
meal_plan_tues = 0
meal_plan_wed = 8
meal_plan_thur = 0
meal_plan_fri = 0
meal_plan_sat = 15
meal_plan_sun = 0

# how may swipes would you need for the full semester?
swipes_only = 3*112

# how many swipes would you need if you also spent money off campus every 
swipes_plus_off_campus = 5*16
swipes_mon = 20
swipes_tues = 23
swipes_wed = 5
swipes_thur = 5
swipes_fri = 13
swipes_sat =  33
swipes_sun =  33

# do you leave for spring break?
leave = True
```

```python
# making pandas df:

if leave == True:
    meal_plan = [0]*112
    meal_plan[0] = meal_plan_cost

    extra_cost_with_meal_plan = [meal_plan_mon, meal_plan_tues, meal_plan_wed, meal_plan_thur, meal_plan_fri, meal_plan_sat, meal_plan_sun]*16
    meal_plan_with_extra = np.add(meal_plan, extra_cost_with_meal_plan)
    meal_plan_with_extra[7*7:7*8] = [0]*7

    buying_swipes_weekly = [0]*112
    for i in range(16):
        buying_swipes_weekly[i*7] = 180 #10 swipes weekly

    buying_swipes_at_start_of_semester = [0]*112
    buying_swipes_at_start_of_semester[0] = np.ceil(swipes_only/10)*90


    buying_some_swipes_weekly = [0]*112
    # for i in range(16):
    #     buying_some_swipes_weekly[i*7] = 90 #10 swipes weekly
    buying_some_swipes_weekly[0] = np.ceil(swipes_plus_off_campus/10)*90
    extra_cost_with_swipes = [swipes_mon, swipes_tues, swipes_wed, swipes_thur, swipes_fri, swipes_sat, swipes_sun]*16
    weekly_swipes_with_extra = np.add(buying_some_swipes_weekly, extra_cost_with_swipes)
    weekly_swipes_with_extra[7*7:7*8] = [0]*7

    data = {
        "meal plan + off campus" : meal_plan_with_extra,
        # "buying weekly swipes" : buying_swipes_weekly,
        "buying all swipes at start of semester" : buying_swipes_at_start_of_semester,
        "buying swipes + off campus" : weekly_swipes_with_extra,
        "meal plan" : meal_plan
    }
    df = pd.DataFrame(data)

else:
    meal_plan = [0]*112
    meal_plan[0] = meal_plan_cost

    extra_cost_with_meal_plan = [meal_plan_mon, meal_plan_tues, meal_plan_wed, meal_plan_thur, meal_plan_fri, meal_plan_sat, meal_plan_sun]*16
    meal_plan_with_extra = np.add(meal_plan, extra_cost_with_meal_plan)

    buying_swipes_weekly = [0]*112
    for i in range(16):
        buying_swipes_weekly[i*7] = 180 #10 swipes weekly

    buying_swipes_at_start_of_semester = [0]*112
    buying_swipes_at_start_of_semester[0] = np.ceil(swipes_only/10)*90


    buying_some_swipes_weekly = [0]*112
    # for i in range(16):
    #     buying_some_swipes_weekly[i*7] = 90 #10 swipes weekly
    buying_some_swipes_weekly[0] = np.ceil(swipes_plus_off_campus/10)*90
    extra_cost_with_swipes = [swipes_mon, swipes_tues, swipes_wed, swipes_thur, swipes_fri, swipes_sat, swipes_sun]*16
    weekly_swipes_with_extra = np.add(buying_some_swipes_weekly, extra_cost_with_swipes)

    data = {
        "meal plan + off campus" : meal_plan_with_extra,
        # "buying weekly swipes" : buying_swipes_weekly,
        "buying all swipes at start of semester" : buying_swipes_at_start_of_semester,
        "buying swipes + off campus" : weekly_swipes_with_extra,
        "meal plan" : meal_plan
    }
    df = pd.DataFrame(data)
```

```python
df.plot(figsize=(14,6))
plt.ylim(-5,500)
plt.legend(loc='upper left', bbox_to_anchor=(0.5, -.1))
plt.figure(figsize=(14, 6))
plt.barh(df.columns, df.sum(), color=['blue', 'orange', 'g', 'r', "purple"])
plt.xticks(rotation=45)
```

The dining plan for spring semester 2024 began on January 6th and ends on April 26th. This time contains 112 days.The 112 days the meal planis active includes 16 days of each day of the week. The 112 days also includes the 5 days from Monday 2/26 to Friday 3/1 when MSU students went on spring break and 1/15 when the university closes for Martin Luther King Jr. Day.
